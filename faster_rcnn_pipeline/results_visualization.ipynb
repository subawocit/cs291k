```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import xml.etree.ElementTree as ET

import matplotlib.pyplot as plt
from PIL import Image

fig, axs = plt.subplots(8, 2, figsize=(8, 30))
ct = 0

for img_id in [657, 1955]:

    tree = ET.parse(f'/hdd/yuchen/pipeline/training_pipeline/data_land4sensor/valid/{img_id}.xml') # Update this path
    root = tree.getroot()
    
    bboxes = []
    for obj in root.findall('.//bndbox'):
        xmin = int(obj.find('xmin').text)
        xmax = int(obj.find('xmax').text)
        ymin = int(obj.find('ymin').text)
        ymax = int(obj.find('ymax').text)
        bboxes.append((xmin, ymin, xmax, ymax))
    
    
    image_width = 224
    image_height = 224

    plt.figure()
    
    
    # Displaying the original image
    img = Image.open(f'/hdd/yuchen/pipeline/training_pipeline/data_land4sensor/valid/{img_id}.jpg')
    axs[ct,0].imshow(img)
    axs[ct,0].set_xlim(0, image_width)
    axs[ct,0].set_ylim(0, image_height)
    axs[ct,0].invert_yaxis()  # Invert Y axis to match the image's origin (top left)
    for (xmin, ymin, xmax, ymax) in bboxes:
        rect = patches.Rectangle((xmin, ymin), xmax-xmin, ymax-ymin, linewidth=3, edgecolor='r', facecolor='none')
        axs[ct,0].add_patch(rect)
    axs[ct,0].set_title('')
    axs[ct,0].axis('off')
    
    img = Image.open(f'/hdd/yuchen/pipeline/training_pipeline/outputs/inference/swin_b_l4s/{img_id}.jpg')
    
    axs[ct,1].imshow(img)
    axs[ct,1].set_title('')
    axs[ct,1].axis('off')

    ct += 1
    
plt.tight_layout()
plt.show()
```
